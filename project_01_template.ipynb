## 4. Answering my Question of Analysis

At the beginning of my report, I chose the variables US sales and Console to closely look at. I will be answering my question of, "Did the console that the video game was released on have an affect on sales between 2004 and 2010?" Using these variables, I will create both a bar chart and a box plot to help us compare them. Using the group by function, I calculated the mean of sales in millions for each console. This is a statistic of our sample. In the box plot, we see that the Wii and Xbox360 had the most sales. This leads me to believe that they were the most popular consoles, and if a game was released on one of them it was more likely to be bought than if a game was released on the other 3 consoles, which was my question of analysis. 

```python
df.groupby('Console')['US Sales (millions)'].mean().plot(kind='bar')
plt.show()
```

```python
df.boxplot(column='US Sales (millions)', by='Console')
plt.show()
```
