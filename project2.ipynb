# Section 3: Visualizing my Data Set

# Section 6: Creating Visualizations

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df_clean = df[["Model Year", "Electric Range"]].dropna()

X = df_clean[["Model Year"]]
y = df_clean[["Electric Range"]]

model = LinearRegression()
model.fit(X, y)

df_clean["Prediction"] = model.predict(X)
df_clean

df_clean.plot.scatter(x="Model Year", y="Electric Range", title="Electric Range Based on Year")

plt.plot(df_clean["Model Year"], df_clean["Prediction"], color='red', linewidth=2)
```

- This graoh shows the data points of the electric ranges of cars throughout the years. I also added a liine of best fit from the Linear Regression model to help show the trends. Although the line is from the model, I do not entirely agree with it. I would think that the line would be positive instead of negative.
